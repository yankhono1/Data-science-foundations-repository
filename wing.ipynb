```python
# Import the necessary libraries 
import tensorflow as tf
from tensorflow import keras
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
```

```python
# load data
data = pd.read_csv('Bird_Data_missing.txt', delimiter = ' ')
data.head()
```

```python
# showing the last 20 birds do not have their species indicated 
data.tail(20)
```

```python
# getting the incomplete records
incomplete_records = data.iloc[148:,[3,4]]
incomplete_records.head(20)
```

```python
# using only the 148 birds with complete data for training
data = data.iloc[:148,:]
data.tail()
```

```python
# extracting features and labels
X = data.loc[:, ['Wing', 'Weight']]
y = data.loc[:, ['Firewing', 'Shadowhawk', 'Thunderbeaks']]
```

```python
X.head(), y.head() 
```

```python
# train-validation-test split
# set 70% of the data for training, the remaining 30% will be for validation and testing
X_train, X_val_test, y_train, y_val_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Split the 30% evenly between validation and testing sets
X_val, X_test, y_val, y_test = train_test_split(X_val_test, y_val_test, test_size=0.5, random_state=42)
```

For this to work define a neural network there has to be an input layer with 2 nodes and hidden layer with 8 nodes and ReLU activation and output layer with 3 nodes and softmax activation

```python
# Define the Neural Network model
model = Sequential()
model.add(Dense(8, input_dim=2, activation='relu'))
model.add(Dense(3, activation='softmax'))
```

```python
# compile the model
# loss- minimises the specifity
# optimizer- see what adjustments can be made (back propigation- quantify how much it should adjust weights and biases to improve the prediction)
# metrics- a performance metric number of porportion of correctly predicted observations
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

Now we train our model. 'epochs' is a hyperparameter that will be set to determine how many times our model will pass through the training set

```python
#train the model
model.fit(X_train, y_train, epochs=50) # reduced epochs for simplicity
```

More complex problems require more epochs, up to a thousand. But it is done so it is not overfitted as we want the general chnage
If accuracy low - optimizer is used for the model to figure out what weights and balances work. when loss is below 1 it is good. 
- our accuarcy=0.6500
- our loss=0.9950
- to determine time we use hyperparameter model

```python
# Predict the class for incomplete records
# list of probabilities for each observation
predicted_probs = model.predict(incomplete_records)
predicted_probs
```

still need to evaluate the performance as well

```python
# argmax returns highest value in the list so it will find the highest probability
predicted_class = np.argmax(predicted_probs, axis=1)
predicted_class
```

```python
# Map this predicted class index to species label
species_labels = ['Firewing', 'Shadowhawk', 'Thunderbeaks']
predicted_species = [species_labels[i] for i in predicted_class]
predicted_species
```

```python
# refining output
print("The predicted classes for the given measurements:")
for i in range(20):
    print(f"{incomplete_records.iloc[i,].values} {predicted_species[i]}")
```
